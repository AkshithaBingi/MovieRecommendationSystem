```python
import pandas as pd

# Creating a mini movie dataset
data = {
    'title': [
        'The Matrix', 'John Wick', 'Avengers: Endgame',
        'Inception', 'Interstellar', 'Iron Man',
        'The Dark Knight', 'Doctor Strange', 'Guardians of the Galaxy',
        'Spider-Man: No Way Home'
    ],
    'genre': [
        'Action Sci-Fi', 'Action Thriller', 'Action Sci-Fi Adventure',
        'Sci-Fi Thriller', 'Sci-Fi Adventure Drama', 'Action Sci-Fi',
        'Action Crime Drama', 'Action Fantasy Sci-Fi', 'Sci-Fi Action Comedy',
        'Action Sci-Fi Superhero'
    ]
}

# Convert to DataFrame and display
df = pd.DataFrame(data)
df.to_csv("movies.csv", index=False)
df.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Load dataset
df = pd.read_csv("movies.csv")

# TF-IDF Vectorization
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['genre'])

# Compute cosine similarity
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Map movie titles to index
indices = pd.Series(df.index, index=df['title']).drop_duplicates()

# Recommendation function
def recommend_movies(title, cosine_sim=cosine_sim):
    if title not in indices:
        return "Movie not found. Please check spelling."
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    return df['title'].iloc[movie_indices]

# Try a recommendation
user_input = "Inception"
print(f"Top recommendations for '{user_input}':")
print(recommend_movies(user_input))
```
